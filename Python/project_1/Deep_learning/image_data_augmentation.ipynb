```python
from keras.preprocessing import image
import matplotlib.pyplot as plt
from keras.preprocessing.image import ImageDataGenerator
```

```python
imagePath = "Deep_learning/data/bird.jpg"
```

```python
img = image.load_img(imagePath,target_size=(500,650))
img = image.img_to_array(img)
type(img)
```

```python
img = img/255.0
```

```python
plt.imshow(img)
```

```python
# create data image generator instance
datagen = ImageDataGenerator( 
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=0.2,
    fill_mode="nearest"
)
```

```python
# Input data should be in batch form
input_batch = img.reshape((1,*img.shape))
input_batch.shape
input_batch
```

```python
i = 0
for output_batch in datagen.flow(input_batch,batch_size= 1):
    plt.figure()
    implot = plt.imshow(image.img_to_array(output_batch[0]))
    i=i+1
    if i == 10:
        break
    plt.axis("off")
    plt.show()
```
